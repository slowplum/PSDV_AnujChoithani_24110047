```python
#Question 1
nums = []
sum_even = 0
product = 1
for i in range(1,21):
    nums.append(i)
for i in nums:
    if i%2 == 0:
        sum_even += i
    else:
        product *= i
mean = sum(nums)/len(nums)
median = (nums[len(nums)//2 - 1] + nums[len(nums)//2])/2
sum_square_for_std_dev = []
for i in nums:
    sum_square_for_std_dev.append((i-mean)**2)
std_div = (sum(sum_square_for_std_dev)/(len(nums) - 1))**0.5

print(sum_even)
print(product)
print(mean)
print(median)
print(std_div)
```

Mutable datatypes are datatypes which can be changed after it has been formed once. For Example: lists in which elements can be appended in the lists or even removed from list. Immutable datatypes are dataypes which can not be changed or altered after they are formed. For example: strings are immutable and it can not be altered like removing or adding character. If we want to alter it we would require to make new string and assigning it. 

```python
#question 3 
import matplotlib.pyplot as plt
n = int(input())
fibo = [1,1]
x = [1,2]
for i in range(3,n+1):
    x.append(i)
if n <= 2:
    print(fibo[0:n])
else:
    while len(fibo) != n:
        fibo.append(fibo[len(fibo)-1] + fibo[len(fibo)-2])
    print(fibo)
    print(x)
plt.plot(x,fibo, label = "fibo",color = 'pink',linewidth = 1.5,marker = '.',markersize = 7)
plt.title("Fibonacci Graph")
plt.xlabel("Nth position")
plt.ylabel("Nth term")
plt.legend()
plt.show()
```

```python
#Question 16
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

set1 = set(eval(input("List1:")))
set2 = set(eval(input("List2:")))
uni = set1.union(set2)
print(union)
inter = set1.intersection(set2)
print(inter)
sym_diff = set1.symmetric_difference(set2)
print(sym_diff)

len1 = len(set1)
len2 = len(set2)
unilen = len(uni)
interlen = len(inter)
symlen = len(sym_diff)

venn2(subsets = (len1,len2,unilen,symlen,interlen), set_labels = ("Set1","Set2","Union","Intersection","Symmetric Difference"))
plt.title("Venn Diagram for Two Sets")
plt.show()
```

```python
#Question 18
def lowerdown(i):
    return i.lower()
f = open('wordcounter.txt',"r")
counter = {}
r = f.readlines()
lines = []
for i in r:
    lines.append(i.rstrip())

for i in lines:
    words = list(map(lowerdown,i.split(' ')))
    for i in words:
        if i in counter:
            counter[i] += 1
        else:
            counter[i] = 1
print(max(counter.values()))
for i in counter:
    if counter[i] == max(counter.values()):
        print(i)

```

```python
#Question 22 
import matplotlib.pyplot as plt
n = int(input("Number of nums you want to enter"))
nums = []
for i in range(n):
    nums.append(int(input("enter number")))
def normal(x):
    x = (x - min(nums))/(max(nums) - min(nums))
    return x

new_nums = list(map(normal,nums))
    
print(new_nums)
plt.plot(nums,new_nums,marker = '*')
plt.show()
```
